```python
# Treating with log
import numpy as np
for col in mushroom.columns:
    if mushroom.skew().loc[col]>0.55:
        mushroom[col]=np.log1p(mushroom[col])
```

```python
mushroom.skew()
```

```python
# Skewness still present, hence treating the skewness with Square Root Transform
list = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-color', 'ring-number', 'ring-type',
       'spore-print-color', 'population', 'habitat']
for i in list:
    mushroom[i]=np.sqrt(np.sqrt(mushroom[i]))
```

# Seperating input and output variable for classification problem

```python
# Now lets seperate input and output variable 
ml_mush=mushroom
ml_mush.shape
```

```python
# defining x variables
# defining x variables
x=ml_mush.iloc[:,1:] 
x.head()
```

```python
# defining y variable
y=ml_mush.iloc[:,:1] 
y.head()
```

```python
x.shape
```

```python
y.shape
```

```python
# Scaling down using standardscaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, auc

max_acc_score=0
for i in range(42,101):
    r_state=i
    x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=r_state,test_size=0.20)
    lgr = LogisticRegression()
    lgr.fit(x_train,y_train)
    y_pred = lgr.predict(x_test)
    acc_score = accuracy_score(y_test,y_pred)
    print("Accuracy score corresponding to random state: ",i," is: ",acc_score)
    if acc_score>max_acc_score:
        max_acc_score=acc_score
        final_r_state=r_state
    print()
    print()
    print("Maximun accuracy score corresponding to ",max_acc_score," is ",final_r_state)
```

```python
# since maximum accuracy score is comming 0.977 in random state  65 we'll check with other models
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=65)
```

```python
# finding best parameters for LogisticRegression using GridSearchCV
from sklearn.model_selection import GridSearchCV
lgr=LogisticRegression()
parameters={'penalty':['l1', 'l2']}
lgrgd=GridSearchCV(lgr,parameters, cv=5)
lgrgd.fit(x_train,y_train)
lgrgd.best_params_
```

```python
# finding best parameters for Support Vector Classification using GridSearchCV
svc=SVC()
parameters={'kernel':('linear','ploy','rbf'), 'C':[1,10]}
svrgd=GridSearchCV(svc,parameters, cv=5)
svrgd.fit(x_train,y_train)
svrgd.best_params_
```

```python
# finding best parameters for KNeighborsClassifier using GridSearchCV
knn=KNeighborsClassifier()
parameters={'n_neighbors':range(1,29)}
knngd=GridSearchCV(knn,parameters, cv=5)
knngd.fit(x_train,y_train)
knngd.best_params_
```

```python
# finding best parameters for RandomForestClassifier using GridSearchCV
rfc=RandomForestClassifier()
parameters={'n_estimators':[100,199]}
rfcgd=GridSearchCV(rfc,parameters, cv=5)
rfcgd.fit(x_train,y_train)
rfcgd.best_params_
```

```python
# finding best parameters for DecisionTreeClassifier using GridSearchCV
dtc=DecisionTreeClassifier()
parameters={'criterion':('gini', 'entropy')}
dtcgd=GridSearchCV(dtc,parameters, cv=5)
dtcgd.fit(x_train,y_train)
dtcgd.best_params_
```

```python
# finding best alpha value for RidgeClassifier using GridSearchCV
rc=RidgeClassifier()
parameters={'alpha':[0.001,0.01,0.1,1]}
rcgd=GridSearchCV(rc,parameters, cv=5)
rcgd.fit(x_train,y_train)
rcgd.best_params_
```

```python
# finding best parameters for gradient boosting using GridSearchCV
gbc=GradientBoostingClassifier()
parameters={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,100,500,1000]}
gbcgd=GridSearchCV(gbc,parameters, cv=5)
gbcgd.fit(x_train,y_train)
gbcgd.best_params_
```

```python
lgr = LogisticRegression(penalty='l2')
svc = SVC(C=1, kernel='rbf')
knn = KNeighborsClassifier(n_neighbors=1)
rfc = RandomForestClassifier(n_estimators=100)
dtc = DecisionTreeClassifier(criterion='gini')
rc  = RidgeClassifier(alpha=1)
gbc = GradientBoostingClassifier(learning_rate=0.01, n_estimators=500)
gnb = GaussianNB()
```

```python
models = []
models.append(('LogisticRegression', lgr))
models.append(('SVC', svc))
models.append(('KNeighborsClassifier', knn))
models.append(('RandomForestClassifier', rfc))
models.append(('DecisionTreeClassifier', dtc))
models.append(('RidgeClassifier', rc))
models.append(('GradientBoostingClassifier', gbc))
models.append(('GaussianNB', gnb))
```

```python
# Testing all required models with best parameters calculated

Model=[]
score=[]
cvs=[]
rocscore=[]
for name,model in models:
    print('********************',name,'**********************')
    print("\n")
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    predict=model.predict(x_test)
    print("\n")
    acc_score=accuracy_score(y_test,predict)
    print("Accuracy Score =", acc_score)
    score.append(acc_score*100)
    print("\n")
    cv_score=cross_val_score(model, x, y, cv=5, scoring="accuracy").mean()
    print("Cross_val_score =", cv_score)
    cvs.append(cv_score*100)
    print("\n")
    print('Classification_report\n', classification_report(y_test,predict))
    print("\n")
    false_positive_rate,true_positive_rate,thresholds=roc_curve(y_test,predict)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print('roc_auc_score',roc_auc)
    rocscore.append(roc_auc*100)
    print("\n")
    cm=confusion_matrix(y_test,predict)
    print(cm)
    print("\n")
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate, label='AUC = %0.2f'% roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

```python
import pandas as pd
result=pd.DataFrame({'Model': Model, 'Accuracy Score': score, 'Cross_val_score': cvs, 'roc_auc_curve': rocscore, })
result
```
